```python
import numpy as np
import pandas as pd
from main_knapsack_functions import *
from logreg_functions import *
```

```python
# 1-constraint dataset (large)
training_1constr = build_training_set(
    num_instances=10000,
    num_items=None,
    instance_generator=generate_knapsack_1constraint
)

# multi-constraint datasets (smaller per m)
dfs_mconstr = []
for num_constr in range(2, 6):
    df_m = build_training_set(
        num_instances=2000,
        num_items=None,
        instance_generator=generate_knapsack_mconstraint,
        m=num_constr
    )
    dfs_mconstr.append(df_m)
training_mconstr = pd.concat(dfs_mconstr, ignore_index=True)

# correlated 1-constraint dataset (random alpha + max_weight per instance)
training_corr = build_training_set(
    num_instances=5000,
    num_items=None,
    instance_generator=correlated_wrapper
)

# combined training set
training_all = pd.concat(
    [training_1constr, training_mconstr, training_corr],
    ignore_index=True
)
```
